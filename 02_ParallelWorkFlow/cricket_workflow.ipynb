```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict
```

```python
class CricketState(TypedDict):

    runs: int
    balls: int
    fours: int
    sixes: int

    strike_rate: float
    balls_per_boundary: float
    boundary_percent: float

    summary: str
```

```python
graph = StateGraph(CricketState)
```

```python
def get_strike_rate(state: CricketState):

    runs = state['runs']
    balls = state['balls']

    return {'strike_rate': runs / balls * 100}

def get_balls_per_boundary(state: CricketState):

    balls = state['balls']
    boundaries = state['fours'] + state['sixes']

    return {'balls_per_boundary': balls / boundaries}

def get_boundary_percent(state: CricketState):

    runs = state['runs']
    fours = state['fours']
    sixes = state['sixes']

    return {'boundary_percent': (fours * 4 + sixes * 6) / runs}

graph.add_node('get_strike_rate', get_strike_rate)
graph.add_node('get_balls_per_boundary', get_balls_per_boundary)
graph.add_node('get_boundary_percent', get_boundary_percent)
```

```python
def get_summary(state: CricketState):
    
    summary = f"""
    Strike Rate: {state['strike_rate']} \nBalls per Boundaries: {state['balls_per_boundary']} \nBoundary percentage: {state['boundary_percent']}
    """

    return {'summary': summary}

graph.add_node('get_summary', get_summary)
```

```python
graph.add_edge(START, 'get_strike_rate')
graph.add_edge(START, 'get_balls_per_boundary')
graph.add_edge(START, 'get_boundary_percent')
graph.add_edge('get_strike_rate', 'get_summary')
graph.add_edge('get_balls_per_boundary', 'get_summary')
graph.add_edge('get_boundary_percent', 'get_summary')
graph.add_edge('get_summary', END)
```

```python
workflow = graph.compile()

initial_state = {
    'runs': 100,
    'balls': 50,
    'fours': 6,
    'sixes': 4
}
workflow.invoke(initial_state)
```
